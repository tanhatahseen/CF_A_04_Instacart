**Preprocessing:** I had made a mistake by filling the missing values in days_since_prior_order column with the median (7), without acknowledging the fact that there is practically no day gap before the 1st order. 

To correct this, I need to make days_since_prior_order = null for every 1st order (where order_number = 1)

```python
#Creating a separate dataframe for the 1st orders, in order to check the number of 1st orders

df_first_order = df[df['order_number']==1]
```

```python
df_first_order.shape
```

```python
#filling the days_since_prior_order with nulls for ordrr_number=1

df.loc[df.order_number == 1, 'days_since_prior_order'] = np.nan
```

```python
#checking the number of nulls after filling the days_since_prior_order with nulls for ordrr_number=1

df.isnull().sum()
```

**Note:** Number of records for order_number = 1 and number of nulls in the days_since_prior_order column after the replacement are equal.
