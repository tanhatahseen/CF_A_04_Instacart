```python
import numpy as np
import pandas as pd
```

```python
df_prods = pd.read_csv('df_prods_clean.csv')
```

```python
#Dropping the index column from the df_prods dataframe

df_prods= df_prods.drop(['Unnamed: 0'], axis=1)
```

```python
df_prods.head()
```

```python
#importing the orders_products_combined dataframe

df_orders_products_combined = pd.read_pickle('/content/drive/MyDrive/orders_products_combined.pkl')
```

```python
df_orders_products_combined.shape
```

The shape of the df_orders_products_combined dataframe is the same as the one I exported from the '4.6: Merging Your Instacart Data' Notebook

```python
#dropping the _merge column from orders_products_combined dataframe

df_orders_products_combined = df_orders_products_combined.drop('_merge', axis=1)
```

```python
df_orders_products_combined.head()
```

```python
# Merging the df_prods and df_orders_products_combined dataframes on product_id

df_orders_products_merged = df_orders_products_combined.merge(df_prods, on = 'product_id')
```
