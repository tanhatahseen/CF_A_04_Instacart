**Step 7:** Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

 

1.   If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
2.   If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
3. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#Creating order_frequency_median column to get the median of day gaps between successive orders of each user id

df['order_frequency_median']=df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df.head()
```

```python
df.loc[df['order_frequency_median'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df.loc[(df['order_frequency_median'] <= 20) & (df['order_frequency_median'] > 10), 'order_frequency_flag'] = 'Regular customer'
df.loc[df['order_frequency_median'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
