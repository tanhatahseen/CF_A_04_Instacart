**Step 4:** Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df.head()
```

```python
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df.head(15)
```

```python
df['loyalty_flag'].value_counts(dropna = False)
```

**Step 5:** Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
df.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max', 'sum']})
```

**Answer:** Seems like, customers who buy frequently from instacart (customers who are labeled as loyal) generally buy less expensive products than customers who are labeled as regular customers. 

**Step 6:** Create a spending flag for each user based on the average price across all their orders using the following criteria:


*   If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
*   If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#Creating mean_price column to get the average expenditure of each user id

df['mean_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df.head()
```

```python
#Creating spending flag for each user based on the mean price

df.loc[df['mean_price'] < 10, 'spending_flag'] = 'Low spender'
df.loc[df['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df.head()
```

```python
df['spending_flag'].value_counts(dropna= False)
```

**Note:** People generally buy cheaper items
