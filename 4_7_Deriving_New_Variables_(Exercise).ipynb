Creating the “price_label” column with the loc() function

```python
df.loc[df['prices']>15,'price_range'] = 'High-range product'
```

```python
df.loc[(df['prices']>5) & (df['prices']<=15),'price_range'] = 'Mid-range product'
```

```python
df.loc[df['prices']<=5,'price_range'] = 'Low-range product'
```

```python
df.head()
```

```python
df['price_range'].value_counts(dropna = False)
```

Creating the “busiest_day” column with For-Loops

```python
df['order_dow'].value_counts(dropna = False)
```

```python
result = []

for value in df["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
df['busiest_day'] = result
```
